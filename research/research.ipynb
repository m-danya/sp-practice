```python
import pandas as pd
import matplotlib
import numpy as np
from ast import literal_eval
```

```python
df_gui = pd.read_csv('df_gui.csv', index_col=0)
df_traces = pd.read_csv('df_traces.csv', index_col=0)

df_gui['clickable_areas'] = df_gui['clickable_areas'].apply(literal_eval)
```

```python
x = []
for i in df_gui['clickable_areas']:
    x.extend(i)
clickable_areas = np.sqrt(pd.Series(x))
```

```python
df_gui
```

```python
p = df_gui['clickable_elements'].plot(kind='hist', figsize=(10, 4),rwidth=0.7, align='left', xlim=(0, 60))

p.set_xlabel('Количество кликабельных элементов')
p.set_title('Распределение количества кликабельных элементов')
p.set_ylabel('');
p.figure.savefig('1.jpg')
```

```python
p = df_traces['length'].plot(kind='hist', figsize=(10, 4),rwidth=0.8, align='left', color='#CC0000')


p.set_title('Распределение длин трасс')
p.set_xlabel('Количество задействованных состояний');
p.set_ylabel('');
p.figure.savefig('2.jpg')
```

```python
p = clickable_areas.plot(kind='hist', figsize=(10, 4),rwidth=0.7, align='left', color='#228B22')

p.set_title('Распределение размера кликабельных элементов')
p.set_xlabel('Средний квадрат из площади разумных кликабельных элементов')
p.set_ylabel('');
p.figure.savefig('3.jpg')
```
