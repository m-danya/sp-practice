```python
import pandas as pd
import matplotlib
import numpy as np
from ast import literal_eval
```

```python
df_gui = pd.read_csv('df_gui.csv', index_col=0)
df_traces = pd.read_csv('df_traces.csv', index_col=0)

df_gui['clickable_areas'] = df_gui['clickable_areas'].apply(literal_eval)
```

```python
x = []
for i in df_gui['clickable_areas']:
    x.extend(i)
```

```python
for i in range(2, 9, 1):
    more = len(df_traces[df_traces['length'] >= i]) / len(df_traces)
    print('Примерно %d%% трасс имеют меньше %d views ' % (100 - more * 100, i))
```

```python
p = df_traces['length'].plot(kind='hist', figsize=(10, 4),rwidth=0.8, align='left', color='#CC0000')

p.set_title('Распределение длин трасс')
p.set_xlabel('Количество скриншотов в трассе');
p.set_ylabel('');
p.figure.savefig('1.jpg')
```

```python
p = df_gui['clickable_elements'].plot(kind='hist', figsize=(10, 4),rwidth=0.7, align='left', xlim=(0, 50), bins=30)

p.set_xlabel('Количество кликабельных элементов в одном gui tree')
p.set_title('Распределение количества кликабельных элементов')
p.set_ylabel('');
p.figure.savefig('2.jpg')
```
